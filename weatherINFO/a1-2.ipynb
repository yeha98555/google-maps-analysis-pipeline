```python
import pandas as pd
#import json
from bs4 import BeautifulSoup
#import xml.etree.ElementTree as ET
import numpy as np
#xmlp = ET.XMLParser(encoding="utf-8")
mon = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
```

```python
def opxml(year = 2015, month = "01"):
    fp = fr'C:\Users\T14 Gen 3\Documents\tibame-project\weatherData\mn_Report_{year}{month}.xml'
    with open(fp, encoding = "UTF-8") as we:
        qw = we.read()
    return BeautifulSoup(qw, "xml")

    
```

```python
def stationN(qwe):
  staN = [np.nan for x in range(4)]
  staN[0] = qwe.find('StationID').text
  staN[1] = qwe.find('StationName').text
  staN[2] = qwe.find('StationNameEN').text
  staN[3] = qwe.find('StationAttribute').text
  # print(pd.DataFrame(staN))
  return staN
```

```python
def maxcou():
    cou = []
    for j in range(2015, 2024):
        for k in mon:
            xdd = opxml(j, k)   
            for asd in xdd.find_all('StationID'):
                if(asd.text not in cou):
                    cou.append(asd.text)
    return len(cou)
```

```python
def hea(ye = 2022, n = 0):
    zxc = [np.nan for x in range(19)]
    xdd = opxml(ye, mon[n])
    zxc[0] = "Year-Month"
    i = 1
    for asd in xdd.find_all('weatherElement'):
        asd = BeautifulSoup(str(asd), 'xml')
        temp = asd.find('tagName').text
        unit = f" ({asd.find('units').text})"
        for qwe in asd.find_all('methodTagName'):
            zxc[i] = f"{qwe.text} {temp}"
            if(("Date" not in qwe.text) and ("Day" not in qwe.text)):
                zxc[i] += unit
            i += 1
    return zxc
```

```python
def ma():
    mm = maxcou()
    ww, hh, dd = 19, 109, mm
    zzxc = [[[np.nan for x in range(ww)] for y in range(hh)] for z in range(dd)]
    ye = 1
    count = 0
    for j in range(2015, 2024):
        for k in mon:
            xdd = opxml(j, k)
            i = count
            for asd in xdd.find_all('location'):
                asd = BeautifulSoup(str(asd), 'xml')
                qaz = asd.find('StationID').text
                #qaz = BeautifulSoup(str(qaz), 'xml')
                bo = 0
                for ll in range(mm):
                    l = zzxc[ll][0][0]
                    #l = BeautifulSoup(str(l), 'xml')
                    if(qaz == l):
                        i = ll
                        bo = 100
                        #print(l.text)
                        continue
                if(bo != 100):
                    i = count
                    zzxc[i][0] = stationN(asd)
                    count += 1
                zzxc[i][ye][0] = f"{str(j)}-{k}"
                for qwe in asd.find_all('AirTemperature'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][1] = qwe.find('Mean').text
                    zzxc[i][ye][2] = qwe.find('Maximum').text
                    zzxc[i][ye][3] = qwe.find('MaximumDate').text
                    zzxc[i][ye][4] = qwe.find('Minimum').text
                    zzxc[i][ye][5] = qwe.find('MinimumDate').text
                for qwe in asd.find_all('Precipitation'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][6] = qwe.find('Accumulation').text
                    zzxc[i][ye][7] = qwe.find('GE01Days').text
                for qwe in asd.find_all('WindSpeed'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][8] = qwe.find('Maximum').text
                    zzxc[i][ye][9] = qwe.find('MaximumDate').text
                for qwe in asd.find_all('WindDirection'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][10] = qwe.find('Maximum').text
                for qwe in asd.find_all('PeakGustSpeed'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][11] = qwe.find('Maximum').text
                    zzxc[i][ye][12] = qwe.find('MaximumDate').text
                for qwe in asd.find_all('PeakGustDirection'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][13] = qwe.find('Maximum').text
                for qwe in asd.find_all('RelativeHumidity'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][14] = qwe.find('Mean').text
                    zzxc[i][ye][15] = qwe.find('Minimum').text
                    zzxc[i][ye][16] = qwe.find('MinimumDate').text
                for qwe in asd.find_all('AirPressure'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][17] = qwe.find('Mean').text
                for qwe in asd.find_all('SunshineDuration'):
                    qwe = BeautifulSoup(str(qwe), 'xml')
                    zzxc[i][ye][18] = qwe.find('Total').text
            ye += 1
    return zzxc
```

```python
def wea(i:int, zzxc = ma()):
    return zzxc[i][:][:]
    # pd.DataFrame(t00, columns = hea())
```

```python
def nam():
    nam = [[np.nan for x in range(4)] for y in range(31)]
    for i in range(31):
        zzxc = wea(i)
        nam[i][0] = zzxc[0][0]
        nam[i][1] = zzxc[0][1]
        nam[i][2] = zzxc[0][2]
        nam[i][3] = zzxc[0][3]
    return nam
```

```python
def datae():
    dt = [[np.nan for x in range(20)] for y in range(3061)]
    o = 0
    for i in range(maxcou()):
        zzxc = wea(i)
        for k in range(108):
            dt[o][0] = zzxc[0][0]
            #print(f"{i}  {k}")
            if(zzxc[k+1][1] != zzxc[k+1][1]):
                continue
            for j in range(19):
                dt[o][j+1] = zzxc[k+1][j]
            o += 1
    #print(o)
    return dt
```

```python
def main():
    zzxc = [[[np.nan for x in range(2)] for y in range(2)] for z in range(2)]
    col = ["ID"]
    col = col + hea()
    zzxc[1] = pd.DataFrame(datae(), columns = col)
    zzxc[0] = pd.DataFrame(nam(), columns=["ID","Name","NameEN","Attribute"])
    return zzxc
```
